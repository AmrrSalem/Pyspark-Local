```python
# Cell 1: Setup PySpark Environment
# Run this cell first in every notebook

import sys
import os

# Configure PySpark to use the correct Python
os.environ["PYSPARK_PYTHON"] = sys.executable
os.environ["PYSPARK_DRIVER_PYTHON"] = sys.executable

# Import PySpark modules
from pyspark.sql import SparkSession
from pyspark.sql.functions import udf, col, when, lit, concat, sum, count, avg, max, min
from pyspark.sql.types import StringType, IntegerType, FloatType, DoubleType, BooleanType

# Create Spark Session
spark = SparkSession.builder \
    .appName("Big Data Lab") \
    .master("local[*]") \
    .config("spark.ui.enabled", "false") \
    .getOrCreate()

# Verify environment
print(f"Python version: {sys.version}")
print(f"Python executable: {sys.executable}")
print(f"PySpark version: {spark.version}")
print("\nSpark session created successfully!")
```

```python
# Cell 2: Example - Create a DataFrame

# Sample data
data = [
    ("Alice", 25, "Engineer"),
    ("Bob", 30, "Manager"),
    ("Charlie", 35, "Analyst"),
    ("David", 28, "Engineer")
]

# Create DataFrame
df = spark.createDataFrame(data, ["Name", "Age", "Role"])

# Show DataFrame
df.show()
```

```python
# Cell 3: Example - UDF (User Defined Function)

# Define a function
def age_category(age):
    if age < 18:
        return "Minor"
    elif age < 30:
        return "Young Adult"
    else:
        return "Adult"

# Create UDF
age_category_udf = udf(age_category, StringType())

# Apply UDF
df_with_category = df.withColumn("AgeCategory", age_category_udf(col("Age")))

# Show result
df_with_category.show()
```

```python
# Cell 4: Example - Group By and Aggregations

# Group by Role and calculate average age
df.groupBy("Role").agg(
    avg("Age").alias("AvgAge"),
    count("*").alias("Count")
).show()
```
